```python
import geopandas
import folium
from shapely.geometry import Point

import gtfs_isochrone
from gtfs_isochrone import main, load
```

```python
gtfs_folder = "../data/orleans/"
data = load.load_prepared_data(gtfs_folder)
```

```python
import datetime

lat = 47.900792
lon = 1.903623
start_datetime = datetime.datetime(2020, 7, 2, 12, 0)
max_duration_seconds = 45 * 60
geojson = gtfs_isochrone.main.compute_isochrone_with_data(data, lat, lon, start_datetime, max_duration_seconds)
```

```python
import folium


m = folium.Map(
    location=[lat, lon],
    zoom_start=12  # Limited levels of zoom for free Mapbox tiles.
)

folium.GeoJson(
    geojson,
    name='geojson'
).add_to(m)

m
```
